**Data transformations**

- Generally data has 3 types of distributions

    - Normal distribution
    
    - postive skewed
    
    - Negative skewed
    
- All the maths concept developed by assumption as data follows normal distribution

- Data skew happens because of many reasons, In that one reason is outliers

- Once we perform the outlier analysis then we can apply transformation methods, if the data does not follows normally

- we have some transformation methods mentioned below

     - Log Transformation
     
     - Exponential Transformation
     
     - Reciprocal Transformation
     
     - Square root Transformation
     
     - Power Transformation
